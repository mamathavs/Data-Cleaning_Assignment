## DATA EXPLORATION USING SEABORN

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
#Import the iris dataset:
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
data=pd.DataFrame(iris.data,columns=iris.feature_names)
label=pd.DataFrame(list(map(lambda x: iris.target_names[x], iris.target)),columns=['Species'])
iris=pd.concat([data,label],axis=1)
print(iris.head())
```

#### 1. Use the displot() to see the distribution of the SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWigthCm features. Plot them as subplots in a single image.

```python
sns.distplot(iris[['sepal length (cm)']], hist=False, rug=True);
sns.distplot(iris[['sepal width (cm)']], hist=False, rug=True);
sns.distplot(iris[['petal length (cm)']], hist=False, rug=True);
sns.distplot(iris[['petal width (cm)']], hist=False, rug=True);
```

#### 2. Do a boxplot of all features except 'Species'

```python
 df = pd.DataFrame(np.random.randn(10,4),
                  columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
boxplot = df.boxplot(column=['sepal length', 'sepal width', 'petal length', 'petal width'])
```

#### 3. Do a countplot for the feature 'Species'

```python
sns.set(style="darkgrid")
ax = sns.countplot(x="Species", data=iris)
plt.show()
```

#### 4. Do a pairplot on the features 'SepalLengthCm','SepalWidthCm','PetalLengthCm', 'PetalWidthCm','Species'

```python
sns.set(style="ticks", color_codes=True)
iris = sns.load_dataset("iris")
g = sns.pairplot(iris)
```

#### 5. Do an Implot on the following SepalLengthCm, PetalLengthCm. Using hue, display the different species in different color

```python
sns.lmplot(x='sepal length (cm)', y='petal length (cm)', data=iris, hue='Species')
```

#### 6. Do a barplot on 'Species' vs 'SepalLengthCm'

```python
sns.barplot(x='Species', y='sepal length (cm)', data=iris)
```

#### 7. Using a heatmap, plot the correlation matrix

```python
sns.heatmap(iris.corr())
```
